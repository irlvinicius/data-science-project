```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('weather.csv', delimiter=',', nrows=100)
weather
```

```python
weather.info()
```

```python
print(weather[['Temperature_c', 'Humidity']].isnull().sum())
```

```python
print(weather[['Temperature_c', 'Humidity']].describe())
```

```python
import seaborn as sns  
sns.scatterplot(weather, x='Humidity', y='Temperature_c')  
```

```python
from sklearn.model_selection import train_test_split

x = weather[['Humidity']].values
y = weather['Temperature_c'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()

linear_regression.fit(x, y)
```

```python
predicao = linear_regression.predict(x)
```

### AUXILIO DE PLOTAGEM (DEEPSEEK)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x=x.flatten(), y=y, color='blue', label='Dados reais')  # x.flatten() para converter para 1D
plt.plot(x, predicao, color='red', linewidth=2, label='Regressão linear')
plt.xlabel('Umidade (%)')
plt.ylabel('Temperatura (°C)')
plt.title('Relação entre Umidade e Temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
